```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.datasets import load_boston
```

```python
ds=load_boston()
```

```python
boston=pd.DataFrame(ds.data, columns=ds.feature_names)
boston.head()
```

```python
boston['Price']=ds.target
boston.head()
```

```python
boston.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X=boston.drop('Price', axis=1)
Y=boston['Price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
```

```python
y_train_predict=lr_model.predict(x_train)
mse=(np.sqrt(mean_squared_error(y_train,y_train_predict)))
print("mse is {}".format(mse))

y_test_predict=lr_model.predict(x_test)
mse2=(np.sqrt(mean_squared_error(y_test,y_test_predict)))
print("mse2 is {}".format(mse2))
```
